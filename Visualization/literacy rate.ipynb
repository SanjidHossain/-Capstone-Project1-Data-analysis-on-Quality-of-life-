```python

def create_dataframe(data):
    columns = ['Country', 'Youth(15- 24) Rate','Youth(15- 24) Year', 'Adult(25+) Rate','Adult(25+) Year', 'Elderly(65+) Rate','Elderly(65+) Year', 'Gender partiality rate', 'Gender partilaity year']
    df = pd.DataFrame([data], columns=columns)
    return df

def add_remaining_rows_to_dataframe(df, table_rows):
    new_rows = []
    for i in range(3, len(table_rows)-1):
        content = [raw_name.text.replace("\n", "").strip() for raw_name in table_rows[i].find_all(name='td')]
        new_rows.append(content)
    df = pd.concat([df, pd.DataFrame(new_rows, columns=df.columns)], ignore_index=True)
    return df

# Example usage
data = ['Afghanistan\u202f*', '47.0', '2011', '31.7', '2011', '20.3', '2011', '0.5', '2011']
df = create_dataframe(data)

# Example remaining rows
remaining_rows = table_rows[3:-1]  # Exclude the last row

df = add_remaining_rows_to_dataframe(df, remaining_rows)
df

```

```python
#erase "*" from the country names
df['Country'] = df['Country'].str.replace('\u202f*', '')
df.dropna(axis=0, how='any',inplace=True)
df
```

```python
#drop rows that has null values
df.dropna(axis=0, how='any',inplace=True)
#save the data to csv file
df.to_csv('temporary_literacy_rate.csv', index=False)
```
