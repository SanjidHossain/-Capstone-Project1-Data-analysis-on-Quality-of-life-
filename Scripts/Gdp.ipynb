```python
# Get the data from the web page
url= "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)"
web_page = bs4.BeautifulSoup(requests.get(url).text, "lxml")
table= web_page.find_all(name="table", attrs={"class":"wikitable"})
table_rows= table[0].find_all("tr")
len(table_rows)
```

```python
def extract_table_data(table_rows, start_row):
    col_names = [raw_name.text.replace("\n", "") for raw_name in table_rows[start_row].find_all(name='td')]
    contents = []

    for row_id in range(start_row + 1, len(table_rows)):
        row_data = table_rows[row_id].find_all(name='td')
        if len(row_data) == len(col_names):
            content = {col_names[idx]: raw_name.text.replace('\n', '').strip() for idx, raw_name in enumerate(row_data)}
            contents.append(content)

    return contents
```
